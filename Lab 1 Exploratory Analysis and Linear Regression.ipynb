## Exploring Data and Linear Regression 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ames = pd.read_csv("Amestrain.csv", error_bad_lines=False)
z = ames['GrLivArea']+ames['BsmtUnfSF']<4000
data = ames[z]
nums0 = data.select_dtypes(include=['int64','float64'])
nums=nums0.dropna()
nums.head(5)
```

# Histograms for Numerical Features

```python
nums.hist(bins=50, figsize=(40,30))
plt.show()
```

# Scatter Plots of Numerical Data vs. Price
I could not get the axis or title labels to display properly, but I labeled each graph manually.  

```python
for i in range(0,len(nums.columns)):
    fig_i, ax_i=plt.subplots()
    ax_i.plot(nums[nums.columns[i]].tolist(),nums.SalePrice,'x', color='Blue')
    ax_i.set_xlabel=nums.columns[i]
    ax_i.set_ylabel=("Sale Price")
    ax_i.set_title=nums.columns[i],"vs Sale Price"
    fig_i.set_size_inches(5,5)
    print(nums.columns[i],'vs Sale Price')
    plt.show()
```

# Linear Regression and Model Prediction
I could not get the axis or title labels to display properly, but I labeled each of the equations.

```python
import sklearn as sk
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
for i in range(0,len(nums.columns)-1):
        fig_i, ax_i=plt.subplots()        
        ax_i.plot(nums[nums.columns[i]].tolist(),nums.SalePrice,'x', color='Blue')
        ax_i.set_xlabel=nums.columns[i]
        ax_i.set_ylabel=("Sale Price")
        ax_i.set_title=nums.columns[i]+"vs Sale Price"
        fig_i.set_size_inches(5,5)
        lin_i=plt.subplot()
        linear_regressor_i=LinearRegression()
        x_i=nums[nums.columns[i]].values.reshape(-1,1)
        y_i=nums.SalePrice.values.reshape(-1,1)
        linear_regressor_i.fit(x_i,y_i)
        price_pred_i=linear_regressor_i.predict(x_i)
        lin_i.plot(x_i,price_pred_i,color='Red')
        lin_i.set_xlabel=nums.columns[i]
        lin_i.set_ylabel=("Sale Price")
        lin_i.set_title=("Sale Price vs", nums.columns[i])
        slope=linear_regressor_i.coef_
        intercept=linear_regressor_i.intercept_
        print(nums.columns[i],'vs Sale Price')
        print('Regression Equation: y=', slope[0][0],'x + ', intercept[0])
        print('RSS:',np.sum((y_i-price_pred_i)**2))
        plt.show()
```
